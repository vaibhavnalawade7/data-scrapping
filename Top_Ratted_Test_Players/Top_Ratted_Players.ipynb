```python
url = 'https://www.cricbuzz.com/cricket-stats/icc-rankings/men/batting'
response = requests.get(url)
content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')

#  Main Div To scrap Data
player_divs = soup.find_all('div', {'class': 'cb-col cb-col-100 cb-rank-tabs'})
```

```python

# List For Storing Data

player_positions = []
player_names = []
player_countries = []
player_ratings = []

for player_name in player_divs:
    try:
        player_position = player_divs.find('div', {'class': 'cb-col cb-col-16 cb-rank-tbl cb-font-16'}).text
        player_positions.append(player_position.text.strip())
    except:
        player_positions.append(np.nan)
        
    try:
        player_name = player_divs.find('a', {'class': 'text-hvr-underline text-bold cb-font-16'}).text
        player_names.append(player_name.text.strip())
    except:
        player_names.append(np.nan)
        
    try:
        player_country = player_divs.find('a', {'class': 'text-hvr-underline text-bold cb-font-16'}).text
        player_countries.append(player_country)
    except:
        player_countries.append(np.nan)
        
    try:
        player_rating = player_divs.find('div', {'class': 'cb-col cb-col-17 cb-rank-tbl pull-right'}).text
        player_ratings.append(player_rating)
    except:
        player_ratings.append(np.nan)
```

```python
player_df = pd.DataFrame ({
    'Position': player_positions,
    'Player Name': player_names,
    'Player Country': player_countries,
    'Player Rating': player_ratings
})
```
